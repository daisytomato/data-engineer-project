```python
def max_number (data):
    timeline =[]
    for e in data:
        if e['event_type'] =='log_in':
            timeline.append((e['timestamp'],1))
        elif e['event_type'] =='log_out':
            timeline.append((e['timestamp'],-1))
    cur_user =0
    max_user =0
    for _, change in timeline:
        cur_user += change
        max_user = max(max_user, cur_user)
    return max_user
        
        
```

```python
def is_valid_ipv4 (s):
    parts = s.split(".")
    for p in parts:
        if not p.isdigit():
            return False
        num = int(p)
        if num <0 or num >255:
            return False
    return True
    
```

```python
print(is_valid_ipv4("192.168.1.1"))     # True
print(is_valid_ipv4("192.168.1.0"))     # True
print(is_valid_ipv4("192.168.100.1.1")) # False
print(is_valid_ipv4("192.1681.1"))      # False
```

```python
def metrics_cal_2(result,metric_name, dim_name, dim_value):
    res_count = 0
    for elements in result:
        if elements[dim_name] == dim_value:
            res_count += elements[metric_name]
    return dim_value, res_count
```

```python
metrics_cal_2(data, "views", "age", "10-20")
```

```python
def top_k (n,data,dim_name,metric_name):
    result=[]
    for row in data:
        result.append(metrics_cal_2(data,metric_name,dim_name,row[dim_name]))
    
    return result
        
         
        
```

```python
def top_3_scores(data):
    dic ={}
    for cat,val in data:
        if cat in data:
            dic[cat] = max(dic[cat], val)
        else:
            dic[cat] = val
    scores = list(dic.values())
    max_scores = sorted(scores,reverse= True)[:3]
    return max_scores
```

```python

class LogEntry:
    def __init__(self, book_id, is_checkout):
        self.book_id = book_id
        self.is_checkout = is_checkout

logs = [
    LogEntry(1, True),
    LogEntry(1, False),
    LogEntry(1, True),
    LogEntry(2, True),
    LogEntry(2, False)
]
```

```python
state ={}
for log in logs:
    book = log.book_id
    checkout = log.is_checkout
    if book in state:
        current_state = state.get(book,False)
        if checkout != current_state:
            return False
        
```
